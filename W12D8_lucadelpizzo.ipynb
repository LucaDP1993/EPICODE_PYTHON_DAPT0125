```python
""" 
Al fine della nostra analisi non abbiamo bisogno di tutte le colonne presenti nel dataset 
quindi tengo solo quelle che ritengo necessarie 
"""


cvd = cvd.loc[:,['iso_code','continent','location','date','total_cases','new_cases','total_deaths','new_deaths','icu_patients','hosp_patients']]
```

```python
""" 
Avendo notato che le righe per cui continent aveva valori nulli corrispondevano a righe che 
contenevano valori aggregati quali European Union o Low-income-countries oppure per England e Wales 
queste righe contenevano solo valori nulli ho deciso di rimuoverle dal dataset
"""


cvd = cvd.dropna(subset = ['continent'])
```

_Per ogni continente trovare il numero di casi dall'inizio della pandemia e la percentuale rispetto al totale mondiale del numero dei casi_

"Per essere più sicuro del risultato ho calcolato il numero totale di casi sia sulla colonna total_cases che su new_cases, ovviamente per quest'ultima bastava una somma mentre per total_cases ho prima raggruppato per città prendendone il valore massimo poi ho sommato questi valori


```python
# Calcolo il numero di casi covid nel mondo

totale_mondiale_new_cases = cvd['new_cases'].sum()
totale_mondiale_tot = cvd.groupby('location')['total_cases'].max().sum()

# Stampo i risultati ottenuti e utilizzo la virgola come separatore delle migliaia
print(f"Totale mondiale casi calcolato su total_cases : {int(totale_mondiale_tot):,}") 
print(f"Totale mondiale casi calcolato su new_cases :{int(totale_mondiale_new_cases):,}")
```

```python
# La variabile africa contiene le righe in cui continent è uguale a Africa

africa = cvd[cvd['continent'] == 'Africa'] 

total_cases_africa = africa.groupby('location')['total_cases'].max().sum()
sum_new_cases_africa = africa['new_cases'].sum()

print(f"Totale casi calcolato su total_cases : {int(total_cases_africa):,}")
print(f"Totale casi calcolato su new_cases : {int(sum_new_cases_africa):,}")
print(f"Percentuale: {(total_cases_africa / totale_mondiale_tot) * 100:.2f}%")
print(f"Percentuale: {(sum_new_cases_africa / totale_mondiale_new_cases) * 100:.2f}%")
```

```python
# La variabile asia contiene le righe in cui continent è uguale a Asia

asia = cvd[cvd['continent'] == 'Asia']

total_cases_asia = asia.groupby('location')['total_cases'].max().sum()
sum_new_cases_asia = asia['new_cases'].sum()

print(f"Totale casi calcolato su total_cases : {int(total_cases_asia):,}")
print(f"Totale casi calcolato su new_cases : {int(sum_new_cases_asia):,}")
print(f"Percentuale: {(total_cases_asia / totale_mondiale_tot) * 100:.2f}%")
print(f"Percentuale: {(sum_new_cases_asia / totale_mondiale_new_cases) * 100:.2f}%")
```

```python
# La variabile europe contiene le righe in cui continent è uguale a Europe

europe = cvd[(cvd['continent'] == 'Europe')]


total_cases_europe = europe.groupby('location')['total_cases'].max().sum()
sum_new_cases_europe = europe['new_cases'].sum()

print(f"Totale casi calcolato su total_cases : {int(total_cases_europe):,}")
print(f"Totale casi calcolato su new_cases : {int(sum_new_cases_europe):,}")
print(f"Percentuale: {(total_cases_europe / totale_mondiale_tot) * 100:.2f}%")
print(f"Percentuale: {(sum_new_cases_europe / totale_mondiale_new_cases) * 100:.2f}%")
```

```python
# La variabile north_america contiene le righe in cui continent è uguale a North America

north_america = cvd[(cvd['continent'] == 'North America')]

total_cases_north_america = north_america.groupby('location')['total_cases'].max().sum()
sum_new_cases_north_america = north_america['new_cases'].sum()

print(f"Totale casi calcolato su total_cases : {int(total_cases_north_america):,}")
print(f"Totale casi calcolato su new_cases : {int(sum_new_cases_north_america):,}")
print(f"Percentuale: {(total_cases_north_america / totale_mondiale_tot) * 100:.2f}%")
print(f"Percentuale: {(sum_new_cases_north_america / totale_mondiale_new_cases) * 100:.2f}%")
```

```python
# La variabile south_america contiene le righe in cui continent è uguale a South America

south_america = cvd[(cvd['continent'] == 'South America')]

total_cases_south_america = south_america.groupby('location')['total_cases'].max().sum()
sum_new_cases_south_america = south_america['new_cases'].sum()

print(f"Totale casi calcolato su total_cases : {int(total_cases_south_america):,}")
print(f"Totale casi calcolato su new_cases : {int(sum_new_cases_south_america):,}")
print(f"Percentuale: {(total_cases_south_america / totale_mondiale_tot) * 100:.2f}%")
print(f"Percentuale: {(sum_new_cases_south_america / totale_mondiale_new_cases) * 100:.2f}%")
```

```python
# La variabile oceania contiene le righe in cui continent è uguale a Oceania

oceania = cvd[(cvd['continent'] == 'Oceania')]

total_cases_oceania = oceania.groupby('location')['total_cases'].max().sum()
sum_new_cases_oceania = oceania['new_cases'].sum()

print(f"Totale casi calcolato su total_cases : {int(total_cases_oceania):,}")
print(f"Totale casi calcolato su new_cases : {int(sum_new_cases_oceania):,}")
print(f"Percentuale: {(total_cases_oceania / totale_mondiale_tot) * 100:.2f}%")
print(f"Percentuale: {(sum_new_cases_oceania / totale_mondiale_new_cases) * 100:.2f}%")
```

```python
"""
Costruisco un Dataframe dove le colonne rappresentano: i continenti e le percentuali 
dei casi per ogni continente sui casi totali
"""

df = pd.DataFrame([
    ['Oceania', total_cases_oceania / totale_mondiale_tot],
    ['Europe', total_cases_europe / totale_mondiale_tot],
    ['North America', total_cases_north_america / totale_mondiale_tot],
    ['South America', total_cases_south_america / totale_mondiale_tot],
    ['Asia', total_cases_asia / totale_mondiale_tot],
    ['Africa', total_cases_africa / totale_mondiale_tot]
], columns=['Continente', 'Percentuale'])
```

```python
# Setto la colonna continente come indice del dataframe
df = df.set_index('Continente')
```

```python
#Realizzo un grafico del DataFrame df, per poter inserire il titolo e altri accorgimenti ho importato la libreria matplotlib.pyplot

df['Percentuale'].plot(kind = 'pie', autopct='%1.1f%%', figsize = (6,6) )
#Tipo di grafico a torta, perchè a mio avviso il migliore per rappresentare la composizione di un intero rispetto alle sue parti
#autopct='%1.1f%%' si usa per inserire le percentuali sul grafico
#figsize = (6,6) per scegliere le dimensioni del grafico


plt.title('Distribuzione percentuale dei casi per continente') # Aggiunge un titolo al grafico 
plt.ylabel('')  # Rimuove l'etichetta dell'asse Y
plt.show()
```

```python
#Creo filtro che mostri solo le righe in cui location è uguale a Germany
filtro_germ = cvd.location.str.contains("Germany",case = False)
cvd[filtro_germ]
```

```python
#Creo filtro che mostri solo le righe in cui location è uguale a France
filtro_fra = cvd.location.str.contains('France',case = False)
cvd[filtro_fra]
```

```python
#Creo start_date e end_date per creare un filtro con un arco di tempo specificato
start_date = '2022-05-01'
end_date = '2023-04-30'
data_range = (cvd.index >= start_date) & (cvd.index <= end_date)
cvd[data_range]
```

```python
#Utilizzo l'operatore logico OR per i filtri sulle nazioni per far si che almeno uno di questi sia verificato e AND per la data che deve verificarsi sempre
filtro_tot = (filtro_ita | filtro_fra | filtro_germ) & data_range
```

```python
#Assegno alla variabile df il dataset risultante dopo l'operazione di masking
df = cvd[filtro_tot]
```

```python
#Trasformo df in una tabella pivot in modo da avere le nazioni come intestazione di colonna e icu_patients come valori della tabella
df_pivot = df.pivot(columns='location',values='icu_patients')
print(df_pivot)
```

```python
#Utilizzo la libreria Seaborn per realizzare il boxplot
sns.boxplot(data=df_pivot)
plt.title('Distribuzione Pazienti ICU')
plt.ylabel('')
plt.xlabel('')

plt.tight_layout()
plt.show()
```

_L'analisi dei boxplot relativi ai pazienti COVID-19 ricoverati in terapia intensiva (ICU) in Italia, Francia e Germania evidenzia differenze significative tra i tre paesi:_

- _Italia: la distribuzione dei pazienti in ICU è piuttosto contenuta, con valori centrali compresi tra 150 e 300. Il minimo si aggira attorno a 100 pazienti, mentre il massimo è vicino a 450. La bassa variabilità e l'assenza di outlier indicano una pressione relativamente stabile sul sistema di terapia intensiva._
- _Francia: mostra valori ICU nettamente superiori, con la maggior parte dei dati distribuiti tra 800 e 1150 pazienti. Il minimo è attorno ai 600, e il massimo si estende fino a 1550. Questo riflette un livello più elevato e variabile di occupazione delle terapie intensive rispetto all’Italia._
- _Germania: registra i valori ICU più alti tra i tre paesi. La distribuzione centrale è compresa tra 750 e 1250 pazienti, con un minimo di circa 500 e un massimo che arriva a 1800. Anche in questo caso non emergono outlier evidenti, ma la variabilità interna è significativa._

*In generale, la pressione sulle terapie intensive è stata molto più elevata in Francia e Germania rispetto all’Italia durante il periodo considerato, segnalando una gestione della pandemia probabilmente influenzata da fattori demografici, sanitari o politici differenti.*

```python
#Assegno alla variabile x il dataset precedente per poter realizzare il grafico
x = cvd[filtro_tot_2021].groupby('location').agg({
    'hosp_patients':'sum'
})
```
